```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'serif'

dataIGCMC = np.loadtxt('IGMC_test_losses.txt', dtype=float)
dataGCMC = np.loadtxt('GCMC_test_losses.txt', dtype=float)
dataGAT = np.loadtxt('MSE_test_losses.txt', dtype=float)[:61] # just the first 61 because IMGC stopped at 61 epochs

fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

x_values = np.arange(len(dataGAT))
ax.plot(x_values, dataIGCMC[:len(x_values)], 'b-', linewidth=2.5, label='IGMC')
ax.plot(x_values, dataGCMC[:len(x_values)], 'r-', linewidth=2.5, label='GCMC')
ax.plot(x_values, dataGAT, 'g-', linewidth=2.5, label='GAT')


ax.set_title("Model Test Loss Comparison", fontsize=16, pad=20)
ax.set_xlabel("Epochs", fontsize=14)
ax.set_ylabel("Loss Value", fontsize=14)
ax.set_ylim(0, 1.5)
ax.set_xlim(0, len(x_values)-1)  # Force start at left edge


ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(framealpha=1, shadow=True)


plt.savefig('loss_comparison.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Check the overfitted model
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'serif'

dataIGCMC = np.loadtxt('GCMC_train_losses.txt', dtype=float)
dataGCMC = np.loadtxt('GCMC_test_losses.txt', dtype=float)

fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

x_values = np.arange(len(dataGAT))
ax.plot(x_values, dataIGCMC[:len(x_values)], 'b-', linewidth=2.5, label='Train')
ax.plot(x_values, dataGCMC[:len(x_values)], 'r-', linewidth=2.5, label='Test')

ax.set_title("GCMC Overfitting", fontsize=16, pad=20)
ax.set_xlabel("Epochs", fontsize=14)
ax.set_ylabel("Loss Value", fontsize=14)
ax.set_ylim(0, 1.5)
ax.set_xlim(0, len(x_values)-1)

ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(framealpha=1, shadow=True)


plt.savefig('loss_comparison.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```
