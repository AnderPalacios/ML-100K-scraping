```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python

sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'serif'


dataIGCMC = np.loadtxt('Huber_test_losses (1).txt', dtype=float)
dataGCMC = np.loadtxt('Huber_test_losses2.txt', dtype=float)
dataGAT = np.loadtxt('Huber_test_losses3.txt', dtype=float)

fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)


x_values = np.arange(len(dataGAT))
ax.plot(x_values, dataIGCMC[:len(x_values)], 'b-', linewidth=2.5, label='Without data')
ax.plot(x_values, dataGCMC[:len(x_values)], 'r-', linewidth=2.5, label='Node data + side info')
ax.plot(x_values, dataGAT, 'g-', linewidth=2.5, label='Node data + side info + movie posters')


ax.set_title("Hubber Loss", fontsize=16, pad=20)
ax.set_xlabel("Epochs", fontsize=14)
ax.set_ylabel("Loss Value", fontsize=14)
ax.set_ylim(0.2, 0.6)
ax.set_xlim(0, len(x_values)-1)


ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(framealpha=1, shadow=True)


plt.savefig('loss_comparison.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'serif'

dataIGCMC = np.loadtxt('MSE_test_losses (1).txt', dtype=float)
dataGCMC = np.loadtxt('MSE_test_losses2.txt', dtype=float)
dataGAT = np.loadtxt('MSE_test_losses3.txt', dtype=float)

fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

x_values = np.arange(len(dataGAT))
ax.plot(x_values, dataIGCMC[:len(x_values)], 'b-', linewidth=2.5, label='Without data')
ax.plot(x_values, dataGCMC[:len(x_values)], 'r-', linewidth=2.5, label='Node data + side info')
ax.plot(x_values, dataGAT, 'g-', linewidth=2.5, label='Node data + side info + movie posters')

ax.set_title("MSE Loss", fontsize=16, pad=20)
ax.set_xlabel("Epochs", fontsize=14)
ax.set_ylabel("Loss Value", fontsize=14)
ax.set_ylim(0.5, 1.1)
ax.set_xlim(0, len(x_values)-1)

ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(framealpha=1, shadow=True)


# 5. Save/show
plt.savefig('loss_comparison.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.titlesize'] = 18

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 8), sharex=True)
plt.tight_layout(pad=4.0)

plot_settings = {
    'linewidth': 2.2,
    'grid_style': {'linestyle': ':', 'alpha': 0.6},
    'legend_settings': {'framealpha': 1, 'shadow': True, 'fontsize': 12}
}

# 1. Huber Loss Plot
x_values = np.arange(len(np.loadtxt('Huber_test_losses3.txt', dtype=float)))
ax1.plot(x_values, np.loadtxt('Huber_test_losses (1).txt', dtype=float)[:len(x_values)],
         'b-', label='Without data')
ax1.plot(x_values, np.loadtxt('Huber_test_losses2.txt', dtype=float)[:len(x_values)],
         'r-', label='Node data + side info')
ax1.plot(x_values, np.loadtxt('Huber_test_losses3.txt', dtype=float),
         'g-', label='Node data + side info + movie posters')
ax1.set_title("Huber Loss Comparison", fontsize=15, pad=15)
ax1.set_ylabel("Loss Value", fontsize=13)
ax1.set_ylim(0.2, 0.6)


# 2. MAE Loss Plot
ax2.plot(x_values, np.loadtxt('MAE_test_losses (1).txt', dtype=float)[:len(x_values)],
         'b-')
ax2.plot(x_values, np.loadtxt('MAE_test_losses2.txt', dtype=float)[:len(x_values)],
         'r-')
ax2.plot(x_values, np.loadtxt('MAE_test_losses3.txt', dtype=float),
         'g-')
ax2.set_title("MAE Loss Comparison", fontsize=15, pad=15)
ax2.set_ylabel("Loss Value", fontsize=13)
ax2.set_ylim(0.6, 1.0)


# 3. MSE Loss Plot
ax3.plot(x_values, np.loadtxt('MSE_test_losses (1).txt', dtype=float)[:len(x_values)],
         'b-')
ax3.plot(x_values, np.loadtxt('MSE_test_losses2.txt', dtype=float)[:len(x_values)],
         'r-')
ax3.plot(x_values, np.loadtxt('MSE_test_losses3.txt', dtype=float),
         'g-')
ax3.set_title("MSE Loss Comparison", fontsize=15, pad=15)
ax3.set_xlabel("Epochs", fontsize=13)
ax3.set_ylabel("Loss Value", fontsize=13)
ax3.set_ylim(0.5, 1.1)

plt.suptitle("Model Performance Comparison Across Different Loss Metrics", y=1.02, fontsize=16)
plt.tight_layout()

plt.savefig('combined_loss_comparison.pdf', format='pdf', bbox_inches='tight', dpi=300)
plt.savefig('combined_loss_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```
